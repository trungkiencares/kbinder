```python
import Magics.macro as magics
```

```python
# Setting the projection
vietnam = magics.mmap(
    subpage_upper_right_longitude= 160.,
    subpage_upper_right_latitude = 40.,
    subpage_lower_left_longitude = 100.,
    subpage_clipping             = 'on',
    subpage_map_projection       = "cylindrical",
    subpage_lower_left_latitude  = 2.,
    page_id_line                 = "off")

# Defining the coastlines
light_background = magics.mcoast(
    map_coastline_sea_shade_colour  = 'white',
    map_coastline_land_shade_colour = 'cream',
    map_grid                        = 'off',
    map_coastline_land_shade        = 'on',
    map_coastline_sea_shade         = 'on',
    map_label                       = 'off',
    map_coastline_colour            = 'tan')

# Load the grib data
wind_from_grib = magics.mgrib(
    grib_input_file_name = './wind10.grib', 
    grib_wind_position_1 = 1,
    grib_wind_position_2 = 2)

# Defining Wind plotting
wind_arrows = magics.mwind(
    legend                = 'on',
    wind_field_type       = 'arrows',
    wind_arrow_thickness  = 1,
    wind_thinning_factor  = 2.5,
    wind_arrow_calm_below = 1,
    wind_arrow_colour     = 'evergreen' )

# Setting the legend
legend = magics.mlegend(
    legend_display_type   = "continuous",
    legend_box_mode       = "automatic",
    legend_box_y_position = 17.00,
    legend_box_x_length   = 32.0,
    legend_box_y_length   = 1.50,
    legend_text_colour    = "charcoal",
    legend_text_font_size = "0.4"
)
```

```python
magics.plot(vietnam, light_background, wind_from_grib, wind_arrows, legend)
```

```python

# Setting the projection
vietnam = magics.mmap(
    subpage_upper_right_longitude= 160.,
    subpage_upper_right_latitude = 40.,
    subpage_lower_left_longitude = 100.,
    subpage_clipping             = 'on',
    subpage_map_projection       = "cylindrical",
    subpage_lower_left_latitude  = 2.,
    page_id_line                 = "off")

# Defining the coastlines
black_background = magics.mcoast(
    map_coastline_sea_shade         =   "on",
    map_coastline_land_shade_colour = "black",
    map_grid                        = "on",
    map_grid_line_style             = "dash",
    map_grid_colour                 = "white",
    map_coastline_land_shade        = "on",
    map_coastline_sea_shade_colour  = "rgb(0.15,0.15,0.15)",
    map_label                       = "on",
    map_coastline_colour            = "tan")

# Load the grib data
wind_from_grib = magics.mgrib(
    grib_input_file_name = './wind10.grib', 
    grib_wind_position_1 = 1,
    grib_wind_position_2 = 2)

# Defining Wind plotting
coloured_wind = magics.mwind(
    legend                                   = 'on',
    wind_field_type                          = 'arrows',
    wind_arrow_unit_velocity                 = 10,
    wind_arrow_min_speed                     = 1.0,
    wind_thinning_factor                     = 2.0,
    wind_advanced_method                     = 'on',
    wind_advanced_colour_selection_type      = 'interval',
    wind_advanced_colour_level_interval      = 5.0,
    wind_advanced_colour_reference_level     = 10.0,
    wind_advanced_colour_max_value           = 25.0,
    wind_advanced_colour_min_value           = 0.25,
    wind_advanced_colour_table_colour_method = 'calculate',
    wind_advanced_colour_direction           = 'clockwise',
    wind_advanced_colour_min_level_colour    = 'turquoise',
    wind_advanced_colour_max_level_colour    = 'purple_red')

# Setting the legend
legend = magics.mlegend(
    legend_display_type   = "continuous",
    legend_box_mode       = "automatic",
    legend_text_colour    = "charcoal",
    legend_text_font_size = "0.4"
)

magics.plot(vietnam, black_background, wind_from_grib, coloured_wind, legend)
```
